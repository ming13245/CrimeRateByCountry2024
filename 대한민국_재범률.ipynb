```python
import pandas as pd

korea_crime_data = pd.read_csv(f"C:/data/project2/대한민국_경찰청_범죄자 범행시 전과 및 재범여부_12_31_2020.csv", 
                         encoding =  'euc-kr')
korea_crime_data.info()
```

```python
new_data = korea_crime_data.drop(columns = ['전과(1범)', '전과(2범)', '전과(3범)',
                                            '전과(4범)', '전과(5범)', '전과(6범)',
                                            '전과(7범)', '전과(8범)', '전과(9범이상)'])
selected_columns = ['범죄중분류', '전과(소계)']
crime_data = new_data[selected_columns]
crime_data_set_index = crime_data.set_index("범죄중분류")
crime_data_set_index
```

```python
murder_data = crime_data_set_index.loc[['살인기수', '살인미수등']]
new_murder_crime = pd.DataFrame({"Count" :[murder_data["전과(소계)"].sum()]}, index = ['살인'])
new_murder_crime
```

```python
rape_data = crime_data_set_index.loc[['강간', '유사강간', '강제추행', '기타 강간 강제추행등']]
new_rape_crime = pd.DataFrame({"Count" :[rape_data["전과(소계)"].sum()]}, index = ['강간'])
new_rape_crime
```

```python
theft_data = crime_data_set_index.loc[['강도']]
new_theft_crime = pd.DataFrame({"Count" :[theft_data["전과(소계)"].sum()]}, index = ['절도'])
new_theft_crime
```

```python
selected_columns_big = ['범죄대분류', '전과(소계)']
big_crime_data = new_data[selected_columns_big]
big_crime_data_set_index = big_crime_data.set_index("범죄대분류")
big_crime_data_set_index
```

```python
big_theft_data = big_crime_data_set_index.loc[['절도범죄']]
new_big_theft_crime = pd.DataFrame({"Count" :[big_theft_data["전과(소계)"].sum()]}, index = ['절도'])
new_big_theft_crime
```

```python
assault_data = crime_data_set_index.loc[['상해', '폭행', '폭력행위등']]
new_assault_crime = pd.DataFrame({"Count" :[assault_data["전과(소계)"].sum()]}, index = ['폭행'])
new_assault_crime
```

```python
result = pd.concat([new_theft_crime, new_big_theft_crime])
final_data = pd.DataFrame({'Count' : [murder_data["전과(소계)"].sum(), rape_data["전과(소계)"].sum(),
                                      result["Count"].sum(), assault_data["전과(소계)"].sum()]})
final_data['Count'] = final_data['Count'].astype(int)
final_data
```

```python
new_index = ['살인', '강간', '절도', '폭행']
final_data.index = new_index
final_data
```
